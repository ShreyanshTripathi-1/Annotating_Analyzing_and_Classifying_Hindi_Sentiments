```python
from collections import defaultdict
d = defaultdict(int)
for i in X_train:
    for _ in i.split():
        if not(_ in stop) and not(_ in Noises):
            d[_] += 1
len(d.keys())
```

```python
import heapq
freq_words = heapq.nlargest(7500, d, key=d.get)
```

```python
freq_words
```

```python
train = []
for _ in X_train:
    vector = []
    t = set(_.split())
    for word in freq_words:
        if word in t:
            vector.append(1)
        else:
            vector.append(0)
    train.append(vector)
train = np.array(train)
train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(train,y_train)
```

```python
Y = []
for _ in X_test:
    vector = []
    t = set(_.split())
    for word in freq_words:
        if word in t:
            vector.append(1)
        else:
            vector.append(0)
    Y.append(vector)
Y = np.asarray(Y)
Y.shape
```

```python
y_pred = gnb.predict(Y)
```

```python
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```

```python
from sklearn.svm import SVC
svm_model_linear = SVC(kernel = 'rbf', C = 1).fit(train,y_train)
svm_predictions = svm_model_linear.predict(Y)
```

```python
count = 0
for i in range(len(svm_predictions)):
    if svm_predictions[i] == y_test.iloc[i]:
        count += 1
```

```python
X = data.iloc[:,0]
y = data.iloc[:,1]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y ,random_state=104,train_size=0.8, shuffle=True)
```

```python
from rank_bm25 import BM25Okapi
corpus = [_.split() for _ in X_train]
```

```python
bm25 = BM25Okapi(corpus)
```

```python
train = [bm25.get_scores(i.split()) for i in X_train]
```

```python
test = [bm25.get_scores(i.split()) for i in X_test]
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0,max_iter=5000)
classifier.fit(train,y_train)
y_pred = classifier.predict(test)
```

```python
count = 0
for i in range(len(test)):
    if y_pred[i] == y_test.iloc[i]:
        count += 1
```

```python
print(count/(i+1)*100)
```

```python
from sklearn.svm import SVC
svm_model_linear = SVC(kernel = 'rbf', C = 1).fit(train,y_train)
svm_predictions = svm_model_linear.predict(test)
```

```python
count = 0
for i in range(len(test)):
    if svm_predictions[i] == y_test.iloc[i]:
        count += 1
print((count/(i+1))*100)
```

```python
from sklearn.svm import SVC
svm_model_linear = SVC(kernel = 'linear').fit(train,y_train)
svm_predictions = svm_model_linear.predict(test)
count = 0
for i in range(len(test)):
    if svm_predictions[i] == y_test.iloc[i]:
        count += 1
print((count/(i+1))*100)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(train,y_train)
y_pred = gnb.predict(test)
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```
